```python
import pandas as pd

url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/car_crashes.csv"

df = pd.read_csv(url)

df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
sns.histplot(df['speeding'])

plt.title("Speeding Distribution")

plt.show()
```

```python
sns.scatterplot(x=df['alcohol'], y=df['speeding'])

plt.title("Alcohol vs Speeding")

plt.show()
```

```python
numeric_df = df.select_dtypes(include=['number'])

plt.figure(figsize=(10,6))

sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')

plt.title("Correlation Heatmap")

plt.show()
```

```python
top_states = df.sort_values(by='total', ascending=False).head(10)

sns.barplot(x=top_states.index, y=top_states['total'])

plt.title("Top States by Total Accidents")

plt.xlabel("State Index")

plt.ylabel("Total Accidents")

plt.show()
```
